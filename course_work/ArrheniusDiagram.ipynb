```python
import matplotlib.pyplot as plt
import numpy as np
import math

def print_graph(T, func):
    fig = plt.figure()

    f = list(map(lambda t: math.log(abs(func(t))) - 1/t, T))
    plt.plot(T, f)

    plt.show()
```

```python
ts = 350
tf = 650
T = np.linspace(ts + 273, tf + 273)
print_graph(T, lambda x : x**2)
```
