```python
import matplotlib.pyplot as plt
import numpy as np
import math

def print_graph(T, func):
    fig = plt.figure()

    f = list(map(lambda t: math.log(abs(func(t))) - 1/t, T))
    plt.plot(T, f)

    plt.show()
```

```python
def print_graph_arr(T, f_arr):
    plt.figure(figsize=(15,5))
    for i in range (T.size):
        f_arr[i] = math.log(abs(f_arr[i])) - 1/(T[i])
    plt.plot(T, f_arr)
    plt.show()
```

```python
ts = 350
tf = 650
T = np.linspace(ts + 273, tf + 273)
print_graph(T, lambda x : x**2)
```
